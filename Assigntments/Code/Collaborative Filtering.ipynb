# Collaborative Filtering

```python
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
```

```python
rating_data = pd.read_csv('ratings.csv')
movie_data = pd.read_csv('movies.csv')
```

```python
rating_data
```

```python
movie_data
```

```python
user_movie_rating = pd.merge(rating_data, movie_data, on='movieId')
user_rating = user_movie_rating.pivot_table('rating', index='userId', columns='title')
movie_rating = user_movie_rating.pivot_table('rating', index='title', columns = 'userId')
user_rating.fillna(0, inplace = True)
movie_rating.fillna(0, inplace=  True)
```

```python
user_movie_rating
```

```python
movie_rating
```

```python
user_rating
```

```python
item_based_CF = cosine_similarity(movie_rating)
item_based_CF = pd.DataFrame(data = item_based_CF, index=movie_rating.index, columns = movie_rating.index)

for i in movie_rating.index:
  item_based_CF[i][i] = 0


item_based_CF
```

```python
user_based_CF = cosine_similarity(user_rating)
user_based_CF = pd.DataFrame(data = user_based_CF, index=user_rating.index, columns = user_rating.index)

for i in user_rating.index:
  user_based_CF[i][i] = 0

user_based_CF
```

```python
def get_item_based_CF(title):
  return item_based_CF[title].sort_values(ascending = False)[:6]
```

```python
get_item_based_CF('Godfather, The (1972)')
```

```python
user_rate = user_rating.values
item_CF = item_based_CF.values
user_CF = user_based_CF.values

item_CF
```

```python
item_based_CF_mat = np.dot(user_rate, item_CF)
user_based_CF_mat = np.dot(user_CF, user_rate)

item_based_CF_df = pd.DataFrame(item_based_CF_mat, index=user_rating.index, columns=user_rating.columns)
user_based_CF_df = pd.DataFrame(user_based_CF_mat, index=user_rating.index, columns=user_rating.columns)
```

```python
user1_user_df = pd.DataFrame.from_dict(user_based_CF_df.loc[2].to_dict(), orient='index', columns = ['rating_expect'])
user1_user_df = user1_user_df.reset_index()
user1_user_df = user1_user_df.rename(columns = {'index':'title'})
user1_user_df = pd.merge(user1_user_df, movie_data, on='title')
user1_user_df = pd.merge(user1_user_df, user1_df, on='movieId')
user1_user_df = user1_user_df[user1_user_df['rating_user']==0]

user1_user_df_sort = user1_user_df.sort_values(by='rating_expect', ascending=False)
user1_user_df_sort.head(10)
```
