# Association Rule


## 1. Data Preprocessing

```python
import pandas as pd
import mlxtend
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
df = pd.read_csv("restaurant_orders.csv")

df
```

```python
df.dropna(inplace=True)
df
```

```python
# order number가 정돈되지않아서 정리

order_dict = {}

lst = df['Order Number'].copy()

cnt = 0
for i in range(len(lst)):
    if order_dict.get(lst[i]) is None:
        order_dict[lst[i]] = cnt
        cnt += 1
    lst[i] = order_dict[lst[i]]
df['Order Number'] = lst
```

```python
df
```

```python
df['Order Number'].unique().shape
```

```python
df_tmp=df[['Order Number','Item Name']]
df_tmp_arr=[[]for i in range(13397)]

for num, i in enumerate(df_tmp['Item Name']) :
    df_tmp_arr[df_tmp['Order Number'][num]].append(i)
```

```python
for num, i in enumerate(df_tmp['Item Name']) :
    df_tmp_arr[df_tmp['Order Number'][num]].append(i)
df_tmp_arr
```

```python
te = TransactionEncoder()
te_ary = te.fit(df_tmp_arr).transform(df_tmp_arr)

df = pd.DataFrame(te_ary, columns = te.columns_)

df
```

##2. Frequent Itemsets Generation

```python
frequent_itemsets = apriori(df,min_support=0.05, use_colnames = True)
frequent_itemsets
```

##3. Association Rule  

```python
association_rules(frequent_itemsets, metric="confidence", min_threshold=0.51)
```

```python
asso_conf = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.1)

asso_lift = asso_conf[asso_conf['lift']>1]

asso_lift = asso_lift.sort_values(by='lift', ascending=False)

asso_lift
```
